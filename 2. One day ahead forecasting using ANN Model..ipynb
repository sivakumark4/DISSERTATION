```python
# Importing necessary libraries
import pandas as pd
import numpy as np
from sklearn.preprocessing import MinMaxScaler
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, LSTM
from tensorflow.keras.callbacks import LearningRateScheduler
import matplotlib.pyplot as plt
from sklearn.metrics import mean_absolute_error, mean_absolute_percentage_error
import os
import pandas as pd
import matplotlib.pyplot as plt
```

# 2. One day ahead forecasting using ANN Model.

### 2.1.One day ahead forecasting using ANN Model for the month of APRIL 2021:

we will use this code to check the accuracy of the ANN model to compare with(ANN,SARIMAX,Holt-Winters model AND LOGISTIC REGRESSION) other models . we are doing this to select the best two models among SARIMA, ANN,SARIMAX AND Holt-Winters model   (NOTE: # we are testing our model and our model will only predict whether next day is a peak day or not. so we will only forecast one day add actual data of that day to train data to retrain)

```python
# Loading the data
data = pd.read_csv('Database_1_capped.csv', parse_dates=['Date and Time'], index_col='Date and Time', dayfirst=True)

# # Forward fill to handle missing values
data = data.ffill() 

# Selecting relevant features
features = ['Demand_Capped', 'Temperature', 'DHI', 'GHI', 'DNI', 'Relative Humidity', 'Wind Speed', 'Holiday', 'Season']
data = data[features]

# Defining the fixed training period
train_start = '2020-04-01'
train_end = '2021-03-31'
train_data = data[train_start:train_end]

# Defining the test period
test_start = '2021-04-01'
test_end = '2021-04-30'
test_data = data[test_start:test_end]

# Data Scaling
scaler = MinMaxScaler(feature_range=(0, 1))
train_scaled = scaler.fit_transform(train_data)

def prepare_data(data, look_back):
    X, y = [], []
    for i in range(look_back, len(data)):
        X.append(data[i-look_back:i, :-1])  # All features except target
        y.append(data[i, 0])  # Target variable (Demand_Capped)
    X, y = np.array(X), np.array(y)
    X = np.reshape(X, (X.shape[0], X.shape[1], data.shape[1] - 1))
    return X, y

# Preparing data for LSTM
look_back = 48  
X_train, y_train = prepare_data(train_scaled, look_back)

# Defining the LSTM model
model = Sequential()
model.add(LSTM(units=100, return_sequences=True, input_shape=(X_train.shape[1], X_train.shape[2]), stateful=False))
model.add(LSTM(units=100, return_sequences=True))
model.add(LSTM(units=50))
model.add(Dense(units=1))

# Compiling the model
model.compile(optimizer='adam', loss='mean_squared_error')

# Defining learning rate scheduler
def lr_scheduler(epoch, lr):
    if epoch > 10:
        return lr * 0.5
    return lr

lr_callback = LearningRateScheduler(lr_scheduler)

# Training the model
model.fit(X_train, y_train, epochs=20, batch_size=32, callbacks=[lr_callback], shuffle=True)

# Daily retraining and forecasting for the entire day (48 intervals of 30 minutes)
daily_forecasts = []
actual_vs_forecast = []
for date in pd.date_range(start=test_start, end=test_end, freq='D'):
    day_forecasts = []
    last_values = data.loc[:date].tail(look_back)
    for interval in range(48):
        # Preparing data for forecasting (using the last 'look_back' values)
        last_values_scaled = scaler.transform(last_values)
        last_values_scaled = np.reshape(last_values_scaled[:, :-1], (1, look_back, last_values.shape[1] - 1))

        # Making a prediction
        forecasted_demand = model.predict(last_values_scaled)
        forecasted_demand_value = forecasted_demand[0][0]

        # Inverse transform the forecasted value to the original scale
        forecasted_demand_value_original = scaler.inverse_transform(
            np.concatenate([np.array([[forecasted_demand_value]]), np.zeros((1, last_values.shape[1] - 1))], axis=1)
        )[0][0]
        day_forecasts.append(forecasted_demand_value_original)

        # Updating last_values with the forecast to predict the next interval
        new_row = pd.DataFrame([[forecasted_demand_value_original] + list(last_values.iloc[-1, 1:].values)], columns=features, index=[last_values.index[-1] + pd.Timedelta(minutes=30)])
        last_values = pd.concat([last_values, new_row]).tail(look_back)

    daily_forecasts.append((date, day_forecasts))
    print(f"Forecasted electricity demand for {date.strftime('%d/%m/%Y')}: {day_forecasts}")

    # Appending actual data to training set and retrain
    actual_values = test_data.loc[date:date + pd.Timedelta(hours=23, minutes=30)]
    if not actual_values.empty:
        new_data = scaler.transform(actual_values)
        train_scaled = np.vstack([train_scaled, new_data])
        X_train, y_train = prepare_data(train_scaled, look_back)
        model.fit(X_train, y_train, epochs=1, batch_size=32, verbose=0, shuffle=True)

    # Storing actual vs forecast for the day
    actual_vs_forecast.extend(list(zip(actual_values['Demand_Capped'], day_forecasts)))

# Creating DataFrame for actual vs forecast and save to CSV
actual_vs_forecast_df = pd.DataFrame(actual_vs_forecast, columns=['Actual', 'Forecast'])
actual_vs_forecast_df.to_csv('actual_vs_forecastANN.csv', index=False)

# Calculating MAD and MAPE
mad = mean_absolute_error(actual_vs_forecast_df['Actual'], actual_vs_forecast_df['Forecast'])
mape = mean_absolute_percentage_error(actual_vs_forecast_df['Actual'], actual_vs_forecast_df['Forecast']) * 100
print(f"Mean Absolute Deviation (MAD): {mad}")
print(f"Mean Absolute Percentage Error (MAPE): {mape}%")

# Plotting the original data and forecasts
plt.plot(data.index, scaler.inverse_transform(scaler.transform(data))[:, 0], label='Original Demand')
for date, forecasts in daily_forecasts:
    forecast_dates = [date + pd.Timedelta(minutes=30 * i) for i in range(48)]
    plt.plot(forecast_dates, forecasts, 'r-', label='Forecasted Demand' if date == daily_forecasts[0][0] else "")
plt.xlabel('Time')
plt.ylabel('Electricity Demand (kW)')
plt.title('One Day Ahead Electricity Demand Forecasting with Daily Updates')
plt.legend()
plt.show()
```

 We got Mean Absolute Deviation (MAD): 539.5798536397931 and Mean Absolute Percentage Error (MAPE): 5.175661737995591% for ANN model's one day ahead forecasting.

### 2.3. Flagging actual and forecasted peak days for ANN one day ahead forecasting models output using 95th percentile approach


we have used the csv with ANN forecasts for 2021 data i.e. actual_vs_forecast_with_datetimeANN.csv for peld flagging using 95th percentile approach

```python
# Loading the dataset
file_path = 'actual_vs_forecast_with_datetimeANN.csv'
df = pd.read_csv(file_path)

# Converting 'Date and Time' column to datetime format
df['Date and Time'] = pd.to_datetime(df['Date and Time'])

# Calculating the 95th percentile for actual and forecasted demand
actual_95th_percentile = df['Actual'].quantile(0.95)
forecast_95th_percentile = df['Forecast'].quantile(0.95)

# Flagging peak days with 1s and non-peak days with 0s for actual and forecasted demand
df['Actual_Peak_Flag'] = (df['Actual'] > actual_95th_percentile).astype(int)
df['Forecast_Peak_Flag'] = (df['Forecast'] > forecast_95th_percentile).astype(int)

# Extracting unique dates by resampling to daily frequency and keeping the maximum flag per day
df['Date'] = df['Date and Time'].dt.date
df_daily = df.resample('D', on='Date and Time').max()
df_daily['Date'] = df_daily.index.date

# Including actual and forecasted demand values in the output CSV
df_daily['Actual_Demand'] = df.resample('D', on='Date and Time')['Actual'].max().values
df_daily['Forecast_Demand'] = df.resample('D', on='Date and Time')['Forecast'].max().values

# Saving the flagged dataset with detailed Date and Time information
df.to_csv('actual_vs_forecast_with_datetimeANN_flagged90.csv', index=False)
df_daily[['Date', 'Actual_Peak_Flag', 'Forecast_Peak_Flag', 'Actual_Demand', 'Forecast_Demand']].to_csv('actual_vs_forecast_with_datetimeANN_flagged_daily999.csv', index=False)

# Counting the number of actual peak days, forecasted peak days, and matching peak days for unique dates
actual_peak_days_count_unique = df_daily['Actual_Peak_Flag'].sum()
forecasted_peak_days_count_unique = df_daily['Forecast_Peak_Flag'].sum()
matching_peak_days_count_unique = ((df_daily['Actual_Peak_Flag'] == 1) & (df_daily['Forecast_Peak_Flag'] == 1)).sum()

# Displaying the results
print(f'Total Unique Actual Peak Days: {actual_peak_days_count_unique}')
print(f'Total Unique Forecasted Peak Days: {forecasted_peak_days_count_unique}')
print(f'Matching Peak Days: {matching_peak_days_count_unique}')
```
